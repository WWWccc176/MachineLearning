```python
### We aim to perform a 1D regression task with following data
###       Input data [0, 1, 2, 3, 4, 5, 6, 7]
###       Output data [0.5, 2.1, 3.8, 6.2, 7.9, 10.4, 11.8, 14.3]
```

```python
### First we need to create corresponding data array with numpy
import numpy as np
x = np.array([0,1,2,3,4,5,6,7])  ### Input as an array in numpy (8,)
y = np.array([0.5,2.1,3.8,6.2,7.9,10.4,11.8,14.3]) ### Output as an array (8,)
### Note that the array you create here is 1D, as a row vector
### In many cases, we will have to make it as a column vector in 2D
```

```python
### Lets check how does the data look like with visualization 
### Matplotlib common python library for visualizaiton
### First import the library
import matplotlib.pyplot as plt

### Using the plot function
### Basic usage: plt.plot(x_axis_data,y_axis_data) 
### The default style of data visualization is line-plot 
### but here we want to visualize discrete points, so we add 'o'
### to make data displayed as circle。
plt.plot(x,y,'o',label='data')
plt.legend() ### this legend commend aims to show the data label on the top left
```

```python
### The data looks very much like linear one, so we perform linear regression
### using ordinary least square method
### Recall that w = (X^T X)^{-1} X^T y
### So we need matrix X and vector y for computation.
### We have already constructed vector y.
### Then for X = [1, x1]
###              [1, x2]
###              [ ... ]
###              [1, xn]
X = np.hstack((np.ones((x.shape[0],1)),x.reshape(-1,1)))
print(X)
### np.ones((a,b)) creates a matrix with size a*b with all the elements being 1
### Here x.shape[0] = 8 means the shape of data (total number of data)
### Therefore we created a (8,1) matrix full of 1.
### x.reshape(-1,1) meaning that we reshape the x (1D row vector) into 2D column vector
```

```python
### Now you have everything, you can start to compute coefficients
w = np.linalg.inv(X.T@X)@X.T@y
print('The coefficients are:',w)
```

```python
### If you want to make prediction with coeffcients, you can do follwing:
### Create test input points ranging from 0 to 7 with 100 points in between
x_test = np.linspace(0,7,100) 
X_test = np.hstack((np.ones((x_test.shape[0],1)),x_test.reshape(-1,1))) ### again form matrix X
y_test = X_test@w
```

```python
### Lets visualize the prediction result
plt.plot(x,y,'o',label='training data')
plt.plot(x_test,y_test,label='prediction with model')
plt.legend()
```
