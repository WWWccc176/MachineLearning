```python
"""
Lab: Build a simple Neural Network (forward only, regression)

Goal:
  - Use NumPy to build a small neural network (MLP)
  - Each layer is a class with random weights and bias
  - Forward prediction only (no training)
  - Compute MSE and MAE losses
"""

# ===================================
# 1) Import the necessary library here
# ===================================
import numpy as np


# ============================================
# 2) Fix random seed so result is reproducible
# ============================================
def set_seed(seed=42):
    np.random.seed(seed)


# ==============================
# 3) Activation functions
# ==============================
def relu(z):
    return np.maximum(0, z)

def tanh(z):
    return np.tanh(z)

def sigmoid(z):
    return 1 / (1 + np.exp(-z))


# ======================================
# 4) Define a Dense layer as a class
# ======================================
class DenseLayer:
    def __init__(self, in_dim, out_dim, activation=None):
        """Create a dense layer with random parameters."""
        self.in_dim = in_dim
        self.out_dim = out_dim
        self.activation = activation

        # Xavier initialization
        limit = np.sqrt(6 / (in_dim + out_dim))
        self.W = np.random.uniform(-limit, limit, (in_dim, out_dim))
        self.b = np.zeros(out_dim)

    def forward(self, x):
        """Compute forward pass through this layer."""
        z = np.dot(x, self.W) + self.b
        if self.activation is None:
            return z
        else:
            return self.activation(z)


# =======================
# 5) Loss functions
# =======================
def mse_loss(y_pred, y_true):
    diff = y_pred - y_true
    return np.mean(diff ** 2)

def mae_loss(y_pred, y_true):
    diff = y_pred - y_true
    return np.mean(np.abs(diff))


# =======================
# 6) Toy regression data
# =======================
def make_regression_data(n=100):
    """
    Generate 1D regression data: y = sin(x) + noise
    x: sampled uniformly from [-π, π]
    y: true value sin(x) with small Gaussian noise
    """
    X = np.linspace(-np.pi, np.pi, n).reshape(-1, 1)
    noise = 0.05 * np.random.randn(n, 1)
    y = np.sin(X) + noise
    return X, y



# =======================
# 7) Main demonstration
# =======================
def main():
    set_seed(123)

    # Generate sample data
    X, y = make_regression_data()

    # Create layers
    layer1 = DenseLayer(1, 5, activation=relu)
    layer2 = DenseLayer(5, 5, activation=relu)
    layer3 = DenseLayer(5, 1, activation=None)  # output layer (linear)

    # Forward pass
    h1 = layer1.forward(X)
    h2 = layer2.forward(h1)
    y_pred = layer3.forward(h2)

    # Compute losses
    mse = mse_loss(y_pred, y)
    mae = mae_loss(y_pred, y)

    print("Predicted y shape:", y_pred.shape)
    print(f"MSE: {mse:.5f}, MAE: {mae:.5f}")


# ===========
# 8) Run it
# ===========
if __name__ == "__main__":
    main()
```
