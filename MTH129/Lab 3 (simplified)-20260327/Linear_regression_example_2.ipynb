```python
### We aim to perform a 1D regression task with following data (same as example 1)
###       Input data [0, 1, 2, 3, 4, 5, 6, 7]
###       Output data [0.5, 2.1, 3.8, 6.2, 7.9, 10.4, 11.8, 14.3]
```

```python
### Same as example 1
import numpy as np
x = np.array([0,1,2,3,4,5,6,7])  ### Input as an array in numpy
y = np.array([0.5,2.1,3.8,6.2,7.9,10.4,11.8,14.3]) ### Output as an array (8,)
```

```python
### Visualize data, same as example 1
import matplotlib.pyplot as plt
plt.plot(x,y,'o',label='data')
plt.legend() ### this legend commend aims to show the data label on the top left
```

```python
### Here we will use sklearn to do ordinary least squre (linear regression)
### Lets import library
import sklearn as sk

### For those machine learning package, typically you only needs to provides input & output for training
### We have defined x and y, but for input x, it needs to be processed.
### In most machine learning package, the input data should be arranged as 
### [num_sample, num_feature], 
### which mean each row representing a data, each column representing a feature of data
### Here we are dealing with 1D linear regression with 8 data 
### The input matrix should looks like [8,1], like a column vector
x = x.reshape(-1,1) 

### Now lets use the model prepared by sklearn
### It is located in the sklearn.linear_model.LinearRegression()
ls_model = sk.linear_model.LinearRegression() 

### Train the model with data x and y
ls_model.fit(x,y)

### lets check the result
print(ls_model.coef_) ### print a in ax+b
print(ls_model.intercept_) ### print b in ax+b
```

```python
### Lets do a test check just like what we did in example 1
### Create test input points ranging from 0 to 7 with 200 points in between
### Dont forget to make into proper shape -> (200,1)
x_test = np.linspace(0,7,200).reshape(-1,1)

### sklearn model has builded a prediction function for you
### you can make prediction by simply add .predict(new_input) to your model
### y_test stores the result
y_test = ls_model.predict(x_test) 
```

```python
### Lets plot it again
plt.plot(x,y,'o',label='data') ### plot data
plt.plot(x_test,y_test,'-',label='OLS prediction') ### plot prediction in range (0,7) completely
plt.legend() ### show label
```
