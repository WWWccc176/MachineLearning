```python
# Example 2 The Gram-Schmidt Process

import numpy as np


x1 = np.array([3, 6, 0])
x2 = np.array([1, 2, 2])

v1 = x1
c1 = np.dot(x2, v1)/np.dot(v1, v1)
v2 = x2 - c1*v1
print(f"An orthogonal basis is {v1, v2}")  
print(f"An orthonormal basis is {(1/np.sqrt(np.dot(v1, v1)))*v1, (1/np.sqrt(np.dot(v2, v2)))*v2}")  
```

```python
# Example 5
import numpy as np

A = np.array([[4, 11, 14],
              [8, 7, -2]])

print(A)
print(f"Dimension of A is {A.shape}")


U, S, VT = np.linalg.svd(A)

Sigma = np.zeros((A.shape[0], A.shape[1]))
Sigma[:len(S), :len(S)] = np.diag(S)

print("A = U S VT, where")


print("U =")
print(U)

print("S = ")
print(Sigma)

print("VT=")
print(VT)

print(f"\nDimension of U is {U.shape}")
print(f"Dimension of S is {Sigma.shape}")
print(f"Dimension of VT is {VT.shape}")



A_reconstructed = U @ Sigma @ VT

print("\n===The product of U, S and VT is ===")
print(A_reconstructed)
```
